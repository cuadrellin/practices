```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA
```

```python
df = sns.load_dataset("iris")
df.iloc[:, :-1]
```

```python
pca_model = PCA(n_components=2)
df_pca = pca_model.fit_transform(df.iloc[:,:-1])
df_pca
```

```python
plt.scatter(df_pca[:,0], df_pca[:,1], c=df["species"].astype("category").cat.codes)
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.show()
```

  CORRELACIÃ“N DE PEARSON: Nos dice quÃ© tan bien se ajustan nuestros datos a una correlaciÃ³n lineal. Si es igual a 1 se ajustan perfectamente a una linea recta con pendiente positiva. Si es igual a -1, los datos se ajustan perfectamente a una lÃ­nea recta con pendiente negativa.
  Cualquier valor entre -1 y 1, nos habla acerca de la pendiente de esta recta y de la varianza que hay en estos datos al momento de ajustar esta recta.
  nos permite ver quÃ© variables se relacionan entre sÃ­ y poder hacer reducciÃ³n de variables.


```python
from scipy.stats import pearsonr
#Calcular la correlaciÃ³n de Pearson
correlacion, _ = pearsonr(df["sepal_length"], df["petal_length"])
print(f"La correlacion de pearson entre sepal length y petal length es: {correlacion}")
```

AnÃ¡lisis de Clusters

```python
from sklearn.cluster import KMeans
```

```python
#Debemos generar el modelo
model = KMeans(n_clusters=3, random_state=42)   #nro de clusters es el nro de clases
df["cluster"] = model.fit_predict(df.iloc[:,:-2])
df
```

```python
sns.scatterplot(data=df,x="sepal_length", y="petal_length", hue="cluster" )
plt.show()
```

DETECCIÃ“N DE OUTLAYERS

```python
#Calcular el Z-score
df["z_score_sepal_length"] = np.abs((df["sepal_length"] - df["sepal_length"].mean())/df["sepal_length"].std())
df
```

```python
outliers = df[df["z_score_sepal_length"] > 2]
outliers
```

```python
sns.scatterplot(data= df, x="sepal_length", y="petal_length", hue="species")
sns.scatterplot(data=outliers, x="sepal_length", y="petal_length", c = "red")
```
